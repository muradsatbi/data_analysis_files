```python
top_groups.revenue = (top_groups.revenue / 1_000_000).apply(lambda x: round(x, 1))
top_groups.profit = (top_groups.profit / 1_000_000).apply(lambda x: round(x, 1))
top_groups['revenue/shop'] = top_groups['revenue/shop'].apply(lambda x: round(x))
```
