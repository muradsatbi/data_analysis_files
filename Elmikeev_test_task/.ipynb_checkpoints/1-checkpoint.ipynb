```python
sales = data['Продажи']
```

```python
sales.head()
```

## Изменим названия, во-первых на латинницу, во-вторых на беспробельные

```python
sales.rename(columns={
    'ИД_Заказа': 'order_id',
    'Ключ_Магазин': 'shop_key',
    'Ключ_Клиент': 'client_key',
    'Ключ_Номенклатура': 'nomenclature_key',
    'Дата заказа': 'order_date',
    'Дата получения': 'receiving_date',
    'Дата отправки': 'shipment_date',
    'Ключ_территория': 'area_key',
    'Количество, шт.': 'amount',
    'Цена, шт': 'price',
    'Полная стоимость': 'total_cost',
    'Себестоимость': 'cost_value',
    'Общая стоимость, руб': 'total_cost_value',
    'Выручка итоговая': 'revenue'    
}, 
             inplace=True)
```

Удалим колонку 'Сидка,%', которая вообще никакой информации в себе не несет

```python
sales.drop(labels='Сидка,%', axis='columns', inplace=True)
```

```python
sales.head()
```

```python
sales.shape
```

```python
sales.dtypes
```

## Изменим все колонки с флоатами, которые по идее инты, на тру инты)

```python
for col in ['order_id', 'shop_key', 'client_key', 'nomenclature_key', 
            'area_key', 'amount']:
    sales[col] = sales[col].apply(int)
```

```python
sales.dtypes
```

```python
sales.head()
```

```python
sales.client_key.nunique()
```

```python
orders_amount = sales.groupby('client_key').agg({'order_id': 'count'}) \
.rename(columns={'order_id': 'orders_amount'}) \
.sort_values(by='orders_amount', ascending=False)
```

```python
orders_amount.head()
```

## Посчитаем, сколько юридических клиентов купило только 1 раз, а сколько больше одного раза

```python
orders_amount['two_or_more'] = orders_amount.orders_amount > 1
```

```python
orders_amount.head()
```

```python
two_or_more = orders_amount.drop(-1, axis='index')
```

```python
orders_amount.drop('two_or_more', axis='columns', inplace=True)
```

```python
two_or_more.head()
```

```python
tom = two_or_more.groupby('two_or_more').agg({'orders_amount': 'count'}) \
.rename(columns={'orders_amount': 'clients'})
```

```python
tom.head()
```

```python
two_or_more.groupby('two_or_more').agg({'orders_amount': 'sum'}) \
.rename(columns={'orders_amount': 'total_orders'})
```

```python
tom = tom.merge(two_or_more.groupby('two_or_more') \
    .agg({'orders_amount': 'sum'}) \
    .rename(columns={'orders_amount':'total_orders'}), left_index=True, 
                right_index=True)
```

```python
tom
```

```python
tom['average_order'] = round(tom.total_orders / tom.clients, 3)
```

```python
tom
```

```python
sales.head()
```

```python
sales_metrics = sales.loc[:, ['client_key', 'area_key', 'amount', 'total_cost_value', 
             'total_cost', 'revenue']]
```

```python
sales_metrics.groupby('client_key').agg({'total_cost_value':'sum',
                                        'total_cost': 'sum',
                                        'revenue':'sum'}).astype(int)
```

```python
orders_amount =orders_amount.merge(sales_metrics.groupby('client_key').agg({'total_cost_value':'sum',
                                        'total_cost': 'sum',
                                        'revenue':'sum'}).astype(int),
                   left_index=True, right_index=True).sort_values(by='revenue',
                                                                 ascending=False)
```

```python
orders_amount.drop(index=-1, inplace=True)
```

```python
orders_amount
```

```python
orders_amount['gain'] = orders_amount.revenue - orders_amount.total_cost_value
```

```python
top_10 = orders_amount.sort_values(by='revenue', ascending=False).head(10)
```

```python
top_10.drop(columns=['orders_amount', 'total_cost_value', 
                     'total_cost', 'total_cost'], inplace=True)
```

```python
top_10.reset_index(inplace=True)
```

```python
top_10
```
